```python
from pandas import read_csv
from matplotlib import pyplot as plt
series = read_csv('/content/^NSEI.csv', header=0, index_col=0, parse_dates=True, squeeze=True)
plt.figure(figsize=(12,5), dpi=100)
series.plot()
plt.title('Stock Market Time series')
plt.show()
```

```python
d = series['Close']
d.plot()
```

```python
D=series['Close'].values
D
```

```python
F=D[0:2000]
```

```python
from PyEMD import EMD
IMF = EMD().emd(F)
N = IMF.shape[0]+1
```

```python
# Plot results
plt.figure(figsize=(12,9))
plt.subplot(N,1,1)
plt.plot(F, 'r')
plt.title("Input signal")
plt.xlabel("Time [s]")
```

```python
for n, imf in enumerate(IMF):
    plt.figure(figsize=(12,9))
    plt.subplot(N,1,n+2)
    plt.plot(imf, 'g')
    plt.title("IMF "+str(n+1))
    plt.xlabel("Time [s]")
```
